```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from transformers import ViTModel
import numpy as np
from sklearn.metrics import accuracy_score
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
from tqdm import tqdm
import os

# Dataset class
class CIFAR10Dataset(Dataset):
    def __init__(self, data, targets=None, transform=None):
        self.images = data
        self.labels = targets
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        label = self.labels[idx] if self.labels is not None else -1
        if self.transform:
            image = self.transform(image)
        return image, label

# Initialize model and preprocessing
device = 'cuda' if torch.cuda.is_available() else 'cpu'
vit_model = ViTModel.from_pretrained("google/vit-base-patch16-224")
vit_model.to(device)
vit_model.eval()

transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# Helper functions
def extract_features(model, dataloader, device):
    all_features = []
    with torch.no_grad():
        for images, _ in dataloader:
            images = images.to(device)
            outputs = model(images)
            features = outputs.last_hidden_state[:, 0, :]
            all_features.append(features.cpu().numpy())
    return np.vstack(all_features)

def compute_prototypes(features, labels):
    prototypes = {}
    for label in np.unique(labels):
        class_features = features[labels == label]
        prototypes[label] = np.mean(class_features, axis=0)
    return prototypes

def classify_with_prototypes(features, prototypes):
    distances = cdist(features, np.array(list(prototypes.values())), metric='euclidean')
    predictions = np.argmin(distances, axis=1)
    return predictions

def save_model_and_prototypes(prototypes, save_path):
    """Save prototypes and other relevant data to a file."""
    torch.save({"prototypes": prototypes}, save_path)
    print(f"Model and prototypes saved to {save_path}")

# Main processing function
def process_dataset(dataset_number, previous_prototypes=None):
    """Process a single dataset and return its prototypes and evaluation accuracy"""

    # Load train data
    train_path = f"/content/{dataset_number}_train_data.tar.pth"
    eval_path = f"/content/{dataset_number}_eval_data.tar.pth"

    train_dict = torch.load(train_path)
    eval_dict = torch.load(eval_path)

    train_images = train_dict['data']
    eval_images = eval_dict['data']
    eval_labels = eval_dict['targets']

    # Create datasets and dataloaders
    train_dataset = CIFAR10Dataset(train_images, transform=transform)
    eval_dataset = CIFAR10Dataset(eval_images, transform=transform)

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=False)
    eval_loader = DataLoader(eval_dataset, batch_size=32, shuffle=False)

    # Extract features
    train_features = extract_features(vit_model, train_loader, device)
    eval_features = extract_features(vit_model, eval_loader, device)

    # If this is not the first dataset, use previous prototypes to get pseudo-labels
    if previous_prototypes is not None:
        train_pseudo_labels = classify_with_prototypes(train_features, previous_prototypes)
        prototypes = compute_prototypes(train_features, train_pseudo_labels)
    else:
        # For the first dataset, use actual labels
        prototypes = compute_prototypes(train_features, train_dict['targets'])

    # Save the current prototypes
    save_model_and_prototypes(prototypes, f"model_prototypes_dataset_{dataset_number}.pth")

    # Make predictions on eval set
    eval_predictions = classify_with_prototypes(eval_features, prototypes)
    accuracy = accuracy_score(eval_labels, eval_predictions)

    return prototypes, accuracy

# Main execution loop
def run_full_evaluation(num_datasets=20):
    """Run evaluation across all datasets and return accuracies"""
    accuracies = []
    current_prototypes = None

    print("Processing datasets...")
    for i in tqdm(range(1, num_datasets + 1)):
        prototypes, accuracy = process_dataset(i, current_prototypes)
        accuracies.append(accuracy)
        current_prototypes = prototypes
        print(f"Dataset {i} Accuracy: {accuracy:.4f}")

    return accuracies

# Run evaluation and plot results
def plot_accuracies(accuracies):
    plt.figure(figsize=(40,40 ))
    plt.plot(range(1, len(accuracies) + 1), accuracies, 'b-o')
    plt.title('Evaluation Accuracy Across Datasets')
    plt.xlabel('Dataset Number')
    plt.ylabel('Accuracy')
    plt.grid(True)
    plt.xticks(range(1, len(accuracies) + 1))
    plt.ylim(0, 1)

    # Add accuracy values on top of points
    for i, acc in enumerate(accuracies):
        plt.text(i + 1, acc + 0.02, f'{acc:.3f}', ha='center')

    plt.savefig('accuracy_plot.png')
    plt.show()

# Run the full evaluation
if __name__ == "__main__":
    print("Starting evaluation across all datasets...")
    accuracies = run_full_evaluation(20)

    # Save accuracies to file
    np.save('dataset_accuracies.npy', accuracies)

    # Plot results
    plot_accuracies(accuracies)

    # Print final summary
    print("\nFinal Summary:")
    print(f"Average Accuracy: {np.mean(accuracies):.4f}")
    print(f"Best Accuracy: {np.max(accuracies):.4f} (Dataset {np.argmax(accuracies) + 1})")
    print(f"Worst Accuracy: {np.min(accuracies):.4f} (Dataset {np.argmin(accuracies) + 1})")
```

```python
def evaluate_model_on_dataset(model_prototypes, dataset_images, dataset_labels):
    """Evaluate a model on a dataset and return accuracy."""
    dataset = CIFAR10Dataset(dataset_images, transform=transform)
    dataloader = DataLoader(dataset, batch_size=32, shuffle=False)

    # Extract features for the dataset
    features = extract_features(vit_model, dataloader, device)

    # Classify using the provided prototypes
    predictions = classify_with_prototypes(features, model_prototypes)

    # Compute and return accuracy
    return accuracy_score(dataset_labels, predictions)


def generate_accuracy_matrix(num_datasets=10):
    """Generate a 10x10 matrix of accuracies."""
    accuracy_matrix = np.zeros((num_datasets, num_datasets))

    print("Generating accuracy matrix...")
    for i in tqdm(range(17, num_datasets + 1)):
        # Load model prototypes for f_i
        model_path = f"model_prototypes_dataset_{i}.pth"
        saved_data = torch.load(model_path)
        model_prototypes = saved_data["prototypes"]

        # Evaluate f_i on D_j for j <= i
        for j in range(1, i + 1):
            dataset_path = f"/content/{j}_eval_data.tar.pth"
            eval_dict = torch.load(dataset_path)
            eval_images = eval_dict["data"]
            eval_labels = eval_dict["targets"]

            # Evaluate the model on the dataset
            accuracy = evaluate_model_on_dataset(model_prototypes, eval_images, eval_labels)

            # Store accuracy in the matrix
            accuracy_matrix[i - 1, j - 1] = accuracy
            print(f"Accuracy of f_{i} on D_{j}: {accuracy:.4f}")
            with open('lacc.txt','a') as f:
              f.write(f"Accuracy of f_{i} on D_{j}: {accuracy:.4f}")
              f.write('\n')

    return accuracy_matrix


# # Generate and save the matrix
if __name__ == "__main__":
    print("Starting evaluation to generate the accuracy matrix...")
    accuracy_matrix = generate_accuracy_matrix(20)

    # Save matrix to a file
    np.save('accuracy_matrix.npy', accuracy_matrix)

    # Print and visualize matrix
    print("\nAccuracy Matrix:")
    print(accuracy_matrix)

    plt.figure(figsize=(40,40))
    plt.imshow(accuracy_matrix, interpolation='nearest', cmap='Blues', vmin=0.9,vmax=1.0)
    plt.colorbar()
    for i in range(accuracy_matrix.shape[0]):
        for j in range(accuracy_matrix.shape[1]):
            plt.text(j, i, f"{accuracy_matrix[i, j]:.4f}",
                     ha="center", va="center", color="black", fontsize=10)
    plt.title('Model-Dataset Accuracy Matrix')
    plt.xlabel('Dataset Index (j)')
    plt.ylabel('Model Index (i)')
    plt.xticks(ticks=np.arange(10), labels=np.arange(1, 11))
    plt.yticks(ticks=np.arange(10), labels=np.arange(1, 11))
    plt.savefig('accuracy_matrix.png')
    plt.show()
```

```python


import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import numpy as np
from torchvision import transforms
import matplotlib.pyplot as plt

# Define a simple Random Autoencoder class
class RandomAutoencoder(nn.Module):
    def __init__(self, input_channels=3):
        super(RandomAutoencoder, self).__init__()
        # Encoder part: Compresses image into a latent space (feature vector)
        self.encoder = nn.Conv2d(input_channels, 16, kernel_size=3, stride=1, padding=1)
        # Decoder part: Reconstructs image from latent space
        self.decoder = nn.ConvTranspose2d(16, input_channels, kernel_size=3, stride=1, padding=1)
        # Noise layers (AdaIN style) to add noise to latent space
        self.noise_linear1 = nn.Linear(1, 16)  # Scaling factor
        self.noise_linear2 = nn.Linear(1, 16)  # Shifting factor

    def forward(self, x):
        encoded = self.encoder(x)
        noise = torch.randn(1, 1)  # Random noise
        noise_scale = F.relu(self.noise_linear1(noise)).view(-1, 16, 1, 1)
        noise_shift = F.relu(self.noise_linear2(noise)).view(-1, 16, 1, 1)
        encoded = encoded * noise_scale + noise_shift
        decoded = self.decoder(encoded)
        return decoded

# Function to generate augmented data
def generate_augmented_data(x_batch, num_augments=3):
    """
    Generates augmented data by applying multiple autoencoders with random noise and
    combining the results with random mixing weights.

    Parameters:
    - x_batch: Input batch of images, shape (batch_size, channels, height, width)
    - num_augments: Number of augmentations to generate for each image

    Returns:
    - Augmented data (torch tensor)
    """
    # Create multiple autoencoders
    autoencoders = [RandomAutoencoder() for _ in range(num_augments)]
    augments = [ae(x_batch) for ae in autoencoders]  # Apply autoencoders to each batch image

    # Randomly combine the original image and the augmented images
    weights = torch.softmax(torch.randn(num_augments + 1), dim=0)  # Random weights
    combined = weights[0] * x_batch  # Original image weight

    # Add augmented images based on weights
    for i, aug in enumerate(augments):
        combined += weights[i + 1] * aug  # Mix augmented data with the original

    return torch.sigmoid(combined)  # Return the augmented images
```

```python
def augment_first_480_images(images, num_to_augment=480, num_augments=3):
    """
    Augments the first 480 images in the batch and returns the augmented images.

    Parameters:
    - images (torch.Tensor): The batch of images to augment, shape (batch_size, 3, 224, 224)
    - num_to_augment (int): The number of images to augment (default 480)
    - num_augments (int): The number of augmentations to generate for each image (default 3)

    Returns:
    - torch.Tensor: A tensor of augmented images, shape (batch_size, 3, 224, 224)
    """
    augmented_images = generate_augmented_data(images[:num_to_augment], num_augments=num_augments)
    # Combine augmented images with the remaining unmodified images
    modified_images = torch.cat((augmented_images, images[num_to_augment:]), dim=0)
    return modified_images
```

```python
def process_dataset_with_augmentation(dataset_number, previous_prototypes=None):
    """Process a single dataset and return its prototypes and evaluation accuracy, including augmentation"""

    # Load train data
    train_path = f"/content/{dataset_number}_train_data.tar.pth"
    eval_path = f"/content/{dataset_number}_eval_data.tar.pth"

    train_dict = torch.load(train_path)
    eval_dict = torch.load(eval_path)

    train_images = train_dict['data']
    eval_images = eval_dict['data']
    eval_labels = eval_dict['targets']

    # Create datasets and dataloaders
    train_dataset = CIFAR10Dataset(train_images, transform=transform)
    eval_dataset = CIFAR10Dataset(eval_images, transform=transform)

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=False)
    eval_loader = DataLoader(eval_dataset, batch_size=32, shuffle=False)

    # Loop over the train loader and augment the first 480 images
    augmented_train_images_list = []
    for images, labels in train_loader:
        # Augment only the first 480 images
        augmented_images = augment_first_480_images(images, num_to_augment=480, num_augments=3)
        augmented_train_images_list.append(augmented_images)

    # Combine the augmented images into one tensor
    augmented_train_images = torch.cat(augmented_train_images_list, dim=0)

    # Use augmented_train_images as the training data for model F_i
    train_features = extract_features(vit_model, augmented_train_images, device)

    # Process the evaluation data as usual
    eval_features = extract_features(vit_model, eval_loader, device)

    # If this is not the first dataset, use previous prototypes to get pseudo-labels
    if previous_prototypes is not None:
        train_pseudo_labels = classify_with_prototypes(train_features, previous_prototypes)
        prototypes = compute_prototypes(train_features, train_pseudo_labels)
    else:
        # For the first dataset, use actual labels
        prototypes = compute_prototypes(train_features, train_dict['targets'])

    # Save the current prototypes
    save_model_and_prototypes(prototypes, f"model_prototypes_dataset_augmented_{dataset_number}.pth")

    # Make predictions on eval set
    eval_predictions = classify_with_prototypes(eval_features, prototypes)
    accuracy = accuracy_score(eval_labels, eval_predictions)

    return prototypes, accuracy
```

```python
def run_full_evaluation_with_augmentation(num_datasets=10):
    """Run evaluation across all datasets and return accuracies, including augmentation"""
    accuracies = []
    current_prototypes = None

    print("Processing datasets with augmentation...")
    for i in tqdm(range(1, num_datasets + 1)):
        prototypes, accuracy = process_dataset_with_augmentation(i, current_prototypes)
        accuracies.append(accuracy)
        current_prototypes = prototypes
        print(f"Dataset {i} Accuracy: {accuracy:.4f}")

    return accuracies
```

```python
def plot_accuracies(accuracies):
    plt.figure(figsize=(12, 6))
    plt.plot(range(1, len(accuracies) + 1), accuracies, 'b-o')
    plt.title('Evaluation Accuracy Across Datasets with Augmentation')
    plt.xlabel('Dataset Number')
    plt.ylabel('Accuracy')
    plt.grid(True)
    plt.xticks(range(1, len(accuracies) + 1))
    plt.ylim(0, 1)

    # Add accuracy values on top of points
    for i, acc in enumerate(accuracies):
        plt.text(i + 1, acc + 0.02, f'{acc:.3f}', ha='center')

    plt.savefig('accuracy_plot_with_augmentation.png')
    plt.show()
```
